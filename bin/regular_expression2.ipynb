```python
##call libraries(re & pandas)
import re
import pandas as pd
from pandas import *

## read a file as table 
gDat = pd.read_table('test2.file', header=None, names = ["v1", "ID", "BIAS", "SPEED", "MORPHWIDTH", "MIDDLINE", "AREA", "loc_x", "loc_y"], delim_whitespace=True)
```

```python
## Extract column(v1)
V1=gDat.v1
## delete column(v1) from table
newgDat = gDat.drop('v1', axis=1)
## delete variable
del gDat
```

```python
a = []
b = []
c = []
e = []
f = []

## looping to get info and put them into dataframe (DATE, PLATE, TIME, STRAIN)
for r in V1:
    M = re.search('./([0-9]{8})_([0-9]{6})/([A-Za-z]+[-]?[0-9]+)', r)
    N = re.search(':([0-9]+[.][0-9]+)', r)
    a = a + [M.group(1)]
    e = M.group (1) + '_' + M.group(2)
    b = b + [e]
    f = f + [N.group(1)]
    c = c + [M.group(3)]
d = {'DATE' : Series(a), 'PLATE' : Series(b), 'TIME' : Series(f), 'STRAIN' : Series(c)}
df = DataFrame(d)
df = df[['DATE', 'PLATE', 'TIME', 'STRAIN']]
```

```python
## delete variables
del a
del b
del c
del d
del e
del f
```

```python
## merge two tables
Newdata = merge(df, newgDat, left_index=True, right_index=True, how='outer')
del newgDat
Newdata
```

```python
## export a merged table to excel files
Newdata.to_excel("newmerge.xlsx", sheet_name='Sheet1')
```
